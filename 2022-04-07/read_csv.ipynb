```python
fpath = "MOCK_DATA.csv"
# Mockaroo - Random Data Generator https://www.mockaroo.com/
```

```python
import pandas as pd
import numpy as np

pd.__version__
```

```python
pd.read_csv(fpath).head()
```

```python
pd.read_csv(fpath).head().info()
```

```python
pd.read_csv(
    fpath, 
    header=None
).head()
```

```python
pd.read_csv(
    fpath, 
    names=list("ABCDEFGHI"),
    index_col="A"
).head()
```

```python
pd.read_csv(
    fpath, 
    usecols=[0,1] # 指定したカラムだけ変換
).head()
```

```python
pd.read_csv(
     fpath,
    dtype={'id': np.float64}
).head()
```

```python
# pyarrow https://pypi.org/project/pyarrow/
pd.read_csv(
     "MOCK_DATA_unique.csv", # 
    engine="pyarrow",
).head()
```

```python
pd.read_csv(
     fpath,
    converters={"last_name":lambda x: x.upper()}
).head()
```

```python
pd.read_csv(
     fpath,
     skiprows=[1,5]
).head()
```

```python
pd.read_csv(
     fpath,
     nrows=3
)
```

```python
pd.read_csv(
     fpath,
     na_values=["Male", "Dav"]
).head()
```

```python
pd.read_csv(
     fpath,
     na_filter=False # データに絶対NAや空白文字が無い場合は、これをFalseにすることでLarge Fileの読み込みが早くなる
).head()
```

```python
pd.read_csv(
     fpath,
     parse_dates=[["date","time"]],
     infer_datetime_format=True, # フォーマットを推論する。もし推論できたらそれに適したパースをするので、場合によっては5-10倍読み込みが早くなる
     keep_date_col=True,
     cache_dates=True, # defaul True, 日付が重複している時などはキャッシュされた日時変換を使用する。読み込みの効率化
).head()
```

```python
pd.read_csv(
     fpath,
     iterator=True,
     chunksize=100,
     )
```

```python
# ‘.gz’, ‘.bz2’, ‘.zip’, ‘.xz’‘.zst’ は読める
pd.read_csv(
    "MOCK_DATA.csv.tar.gz",
).head()
```
